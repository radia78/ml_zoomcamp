# EDA

```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the data of amazon product reviews
df = pd.read_csv('ratings_Beauty.csv')
```

```python
print(df.head())
```

```python
# See the number of users given for each rating
df.Rating.apply(lambda x:int(np.ceil(x))).value_counts().plot.bar(figsize=(10,6))
plt.title('Number of ratings given by users')
plt.xlabel('Ratings')
plt.ylabel('Count of ratings')
plt.show()
```

From the data we see that most users will usually give a rating of 5 for most products. This indicates that most users usually give a 5 if they like the product and usually a 1 or 2 for products that they don't like.

```python
# Ratings each day
df['Timestamp'] = df['Timestamp'].map(lambda x: dt.fromtimestamp(x))
print(df['Timestamp'].describe(datetime_is_numeric=True))

df['Timestamp'].dt.day_name().value_counts().plot.bar(color='#FF6631',figsize=(10,6))
plt.title("Review counts (By day name)",fontweight='bold',fontsize=17)
plt.xlabel("Day")
plt.ylabel("User count")
plt.show()
```

Looking at the chart, it looks like that most reviews are done during the weekdays and sunday. This might indicate that most of these reviews are part of work as Friday is usually a day where people leave work early to prepare for the weekends and Sunday is a day where people get ready for work on Monday.

```python
# Ratings by the hour
df.Timestamp.dt.hour.value_counts().plot.bar(figsize=(10,6))
plt.xlabel("Hour (24)")
plt.ylabel("Review count")
plt.title("Reviews by hour",fontweight='bold')
plt.show()
```

From the graph, it looks like that reviews are usually done between the hours of 7:00 PM and 8:00 PM.

From these charts, I'm going to create a new feature to determine if the review is during the work week or weekend.

# Feature Selection

We require two types of features. First is feature by user and the second is feature by product.

```python
def get_model_input(X_u, X_m, y, tgt_users):

    merged = pd.merge(X_u, y, on=['UserId'], how='inner')
    merged = pd.merge(X_m, merged, on=['ProductId'], how='outer')
    merged = merged.query('UserId in @tgt_users')

    merged.fillna(0, inplace=True)
    features_cols = list(merged.drop(columns=['UserId', 'ProductId', 'Rating', 'Timestamp']).columns)

    query_list = merged['UserId'].value_counts()

    merged = merged.set_index(['UserId', 'ProductId'])

    query_list = query_list.sort_index()

    merged.sort_index(inplace=True)

    df_x = merged[features_cols]

    df_y = merged['Rating']
    
    return df_x, df_y, query_list
```
