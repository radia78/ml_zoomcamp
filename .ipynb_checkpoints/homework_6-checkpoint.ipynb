### Data Preparation

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split

# Load-in the data
df = pd.read_csv('homework_data/housing.csv')

# Fill in the NA's with 0
df.fillna(0, inplace=True)

# Apply the log transformation to median_house_value
df['median_house_value'] = np.log1p(df['median_house_value'])

# Use exclusively these columns
X = df[[
        'latitude', 
        'longitude', 
        'housing_median_age', 
        'total_rooms',
        'total_bedrooms',
        'population',
        'households',
        'median_income',
        'ocean_proximity'
]]

y = df.median_house_value

# Use dict_vectorizer to turn dataframes into matrices
Xdicts = X.to_dict(orient='records')
dv = DictVectorizer(sparse=False)
X = dv.fit_transform(Xdicts)

# Split the data accordingly
Xtr, Xdev, ytr, ydev = train_test_split(X, y, test_size=0.4, random_state=1)
Xval, Xtest, yval, ytest = train_test_split(Xdev, ydev, test_size=0.5, random_state=1)
```

### Question 1

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_text

# Train the model
clf1 = DecisionTreeRegressor(max_depth=1)
clf1.fit(Xtr, ytr)

# Find which feature is used for splitting the data
print(export_text(clf1, feature_names=dv.get_feature_names()))
```

### Question 2

```python
from sklearn.ensemble import RandomForestRegressor

# Train the model
clf2 = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
clf2.fit(Xtr, ytr)

# Evaluate the model
ypred_clf2 = clf2.predict(Xval)
rmse = np.sqrt(np.sum((ypred_clf2 - yval)**2/len(yval)))
print(rmse)
```

### Question 3

```python
# Find the optimal n_estimators
n_est = np.arange(10, 200, 10)
rmse_est = []

for e in n_est:
    clf_temp = RandomForestRegressor(n_estimators=e, random_state=1, n_jobs=-1)
    clf_temp.fit(Xtr, ytr)
    ypred_temp = clf_temp.predict(Xval)
    rmse = np.sqrt(np.sum((ypred_temp - yval)**2/len(yval)))
    rmse_est.append(rmse)
    
rmse_est = np.array(rmse_est)
```

```python
# PLot the graph
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(n_est, rmse_est)
plt.title("RMSE for each N_estimators using RandomForestRegressor")
plt.xlabel("N_estiamtors")
plt.ylabel("RMSE")
plt.show()
```

It looks like that the RMSE stops improving after 150 estimators

### Question 4

```python
# Create the list of max_depth
depths = np.array([10, 15, 20, 25])
rmse_depth = {}
for e in n_est:
    for d in depths:
        clf_temp = RandomForestRegressor(n_estimators=e, max_depth=d, random_state=1, n_jobs=-1)
        clf_temp.fit(Xtr, ytr)
        ypred_temp = clf_temp.predict(Xval)
        rmse = np.sqrt(np.sum((ypred_temp - yval)**2/len(yval)))
        rmse_depth['{depth}:{nest}'.format(depth=d, nest=e)] = rmse
```

```python
print(min(rmse_depth, key=rmse_depth.get))
```

### Question 5

```python
# Train the specific model
clf3 = RandomForestRegressor(n_estimators=10, max_depth=20, random_state=1, n_jobs=-1)
clf3.fit(Xtr, ytr)

# Find the feature importance and assign the feature name
feature_names = dv.get_feature_names_out()
feature_importance = np.float64(clf3.feature_importances_)
feature_dict = {feature_names[i]: feature_importance[i] for i in range(len(feature_importance))}

# Show the most important feature
print(max(feature_dict, key=feature_dict.get))
```

### Question 6

```python
import xgboost as xgb

# Turn data into DMatrix in XGBoost
feature_names[5] = 'ocean_proximity=LESS 1H OCEAN'
dtrain = xgb.DMatrix(Xtr, label=ytr, feature_names=feature_names)
dval = xgb.DMatrix(Xval, label=yval, feature_names=feature_names)

# Create a watchlist for the algorithm
watchlist = [(dtrain, 'train'), (dval, 'validation')]

# Create parameters for the model
xgb_params = {
    'eta': 0.3, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

# Train the model
model = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist)
```

```python
# Create parameters for the model
xgb_params2 = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

# Train the model
model2 = xgb.train(xgb_params2, dtrain, num_boost_round=100, evals=watchlist)
```

```python
# Create parameters for the model
xgb_params3 = {
    'eta': 0.01, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

# Train the model
model3 = xgb.train(xgb_params3, dtrain, num_boost_round=100, evals=watchlist)
```

Judging by the result from the watchlist of each model, I think eta=0.3 is the best out of all of them. However, keep in mind that when eta becomes smaller, it has to increase the number boost rounds since training rate is slower.
