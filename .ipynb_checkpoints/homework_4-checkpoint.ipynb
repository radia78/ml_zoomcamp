### Question 3

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# Compute threshold array and probabilities
threshold = np.arange(0, 1, 0.01)
yprob = clf.predict_proba(Xval)[:, 1]

# Compute precision and recall
p = []
r = []

for t in threshold:
    pred = np.array(yprob >= t).astype(int)
    precision = precision_score(yval, pred)
    recall = recall_score(yval, pred)
    p.append(precision)
    r.append(recall)

# Convert list to numpy array
p = np.array(p)
r = np.array(r)
```

```python
# Plot both lines
plt.plot(threshold, p, label = 'Precision')
plt.plot(threshold, r, label = 'Recall')
plt.legend()
plt.show()
```

From the plot, it looks like the curves intersect when the threshold is equal to 0.18

### Question 4

```python
# Compute f1 score
f1 = 2*(p*r)/(p + r)

# Plot the f1 curve
plt.plot(threshold, f1, label='F1 Score')
plt.legend()
plt.show()
```

Judging from the graph above, the graph is maximal near 0.2
